## Motivation and Problem Statement

Energy consumption and greenhouse gas emission has been an ongoing battle in order to save our environment. Based on the 2017 data in Seattle, the [Built Environment was the primary contributor to Green House Gas emission](https://your.kingcounty.gov/dnrp/climate/documents/201907-KingCounty-GHG-Emissions-Analysis.pdf). Thus, the city has been coming up with various measures to reduce the Green House Gas emission. The city  has one of the more stringent building energy codes among the others as part of the city's plan to become [carbon neutral by 2050](https://www.seattle.gov/council/meet-the-council/mike-obrien/carbon-neutral#:~:text=In%202011%2C%20the%20City%20Council,transportation%2C%20building%20energy%20and%20waste.). As a result, the City has been requiring non-residential and multifamily buildings to report their annual energy performance. The data that has been made available to public allow us to analyze every building larger than 20,000 sqf in gross foot area within Seattle. 

Overall, this project aims to identify how buildings in Seattle have been able to contribute to Seattle's 2050 plan by reducing their greenhouse gas emission. At the end of this project it is expected that we can understand the average energy performance profile of each building type in Seattle. 

**Question 1**: *Which building type had been able to reduce their green gas emissions between 2015-2019?*

***Question 2:** *How did Seattle nighborhoods perform in terms of energy consumption and green house gas emission within the 2015-2019 time period?*

## Greenhouse Gas Overview

Accoring to EPA, greenhouse gases are gases that trap heat in the atmosphere (EPA, 2019). Carbon dioxide, Methane, Nitrous Oxide, and Fluorinated gases are classified greenhouse gases because of their potential to cause climate change through their heat-trapping properties. There areseveral sources of GHG but the most notable ones are from transportation and buildings (*see fig.1*). Buildings produces GHG through their energy consumptions like cooking, space, and water heating. According to the City of Seattle, [86% of these activities are largely](https://www.seattle.gov/environment/climate-change/climate-planning/performance-monitoring) powered by the fossil gas. The City of Seattle reported tat GHG emission has increased by 8.3% in the past year. Building codes and more regulations were developed to curb the increase, such as utilization of more environmentally friendly or locally sourced materials.

## Prior Studies and Hypotheses

The City of Seattle has developed a [Seattle Climate Action](http://greenspace.seattle.gov/wp-content/uploads/2018/04/SeaClimateAction_April2018.pdf) in 2018. This resolution aims to reduce the energy emissions from building by 38% in 2030. Within the resolution, the City mandated a benchmarking for non-residential buildings to report their energy usage and GHG emissions annually. This allows the City to track the buildings' energy usage over the years. However, there has been minimum detail on which building types produces more GHG. By identifying the GHG profile, it is expected that we can plan a more focused effort in reducing it.


*Hypotheses:* With the requirements in place, it is expected that the GHG emissions can be reduced over the years.
