## 7. Findings

### 7.1. Total GHG Emission from 2015 to 2019

First, the csv files are loaded into the notebook. There are 6 files that represents each year from 2015 - 2019

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df1 = pd.read_csv('2015_Building_Energy_Benchmarking.csv')
df2 = pd.read_csv('2016_Building_Energy_Benchmarking.csv')
df3 = pd.read_csv('2017_Building_Energy_Benchmarking.csv')
df4 = pd.read_csv('2018_Building_Energy_Benchmarking.csv')
df5 = pd.read_csv('2019_Building_Energy_Benchmarking.csv')
```

Once the reports have been loaded, I attempted to print one set of data to see if the process had been successful

```python
#print the 1st cv to see if it has been successfully uploaded
print (df1)
```

The next step is to find out how many building managers filed their energy benchmark for each year. Fortunately, each line in the .csv files corresponds to one report. Then, the data are printed to test if the code is running

```python
report2015=len(df1)
report2016=len(df2)
report2017=len(df3)
report2018=len(df4)
report2019=len(df5)
print ("There are", report2015, " reports in 2015")
print ("There are", report2016, " reports in 2016")
print ("There are", report2017, " reports in 2017")
print ("There are", report2018, " reports in 2018")
print ("There are", report2019, " reports in 2019")
```

The data suggest that there were more reports in 2018 than other years.  

```python
sumdf1 = df1["GHGEmissions(MetricTonsCO2e)"].sum()
sumdf2 = df2["TotalGHGEmissions"].sum()
sumdf3 = df3["TotalGHGEmissions"].sum()
sumdf4 = df4["TotalGHGEmissions"].sum()
sumdf5 = df5["TotalGHGEmissions"].sum()

print ("Total GHG emission in 2015 is ", sumdf1)
print ("Total GHG emission in 2016 is ", sumdf2)
print ("Total GHG emission in 2017 is ", sumdf3)
print ("Total GHG emission in 2018 is ", sumdf4)
print ("Total GHG emission in 2019 is ", sumdf5)
```

A new dataset that compiles the annual total emission for all buildings that reported their GHG emission. The result indicated that as the year progresses, buildings around Seattle produce more GHG emission. The result is understandable since more buildings filed the report in more recent years. 
Once the data has been compiled into a table, it can plotted to create a line chart.

```python
ghgemission = {'Year':[2015,2016,2017,2018,2019],'GHG_Emission':[sumdf1, sumdf2, sumdf3, sumdf4, sumdf5]}
totalemission = pd.DataFrame(ghgemission,columns=['Year','GHG_Emission'])
print (totalemission)
```

The line chart below demonstrated the increase of GHG emission. It can be inferred that between  2017 and 2018, the emission increased by almost twice. 

```python
totalemission.plot(x = 'Year', y = 'GHG_Emission', kind = 'line')
plt.savefig('total_emission')
```

Counting total emission of the buildings shows the increase of GHG emission. While the visualization indicated the increase in GHG emission across Seattle, we still don't know the average performance of each building. In the next subsection, the average GHG emission is be measured.

### 7.2. Average GHG Emission from 2015 to 2019

The process is rather similar to the previous subsection, except this time the mean is calculated. 

```python
meandf1 = df1["GHGEmissions(MetricTonsCO2e)"].mean()
meandf2 = df2["TotalGHGEmissions"].mean()
meandf3 = df3["TotalGHGEmissions"].mean()
meandf4 = df4["TotalGHGEmissions"].mean()
meandf5 = df5["TotalGHGEmissions"].mean()

print ("Total GHG emission in 2015 is ", meandf1)
print ("Total GHG emission in 2016 is ", meandf2)
print ("Total GHG emission in 2017 is ", meandf3)
print ("Total GHG emission in 2018 is ", meandf4)
print ("Total GHG emission in 2019 is ", meandf5)
```

The following code compiles the average emission into a table

```python
## create new data
ghgemission = {'Year':[2015,2016,2017,2018,2019],'Average_Emission':[meandf1, meandf2, meandf3, meandf4, meandf5]}
averageemissiontable = pd.DataFrame(ghgemission,columns=['Year','Average_Emission'])
print (averageemissiontable)
```

The table is plotted into a line chart, similar to previous section.

```python
averageemissiontable.plot(x = 'Year', y = 'Average_Emission', kind = 'line')
plt.savefig('average_emission_table')
```

It turns out, average GHG emission increased significantly in 2019. This means, on average, buildings produce more GHG emission in recent years. If 2019 data is excluded, average GHG emission had been relatively the same from 2015 to 2019
